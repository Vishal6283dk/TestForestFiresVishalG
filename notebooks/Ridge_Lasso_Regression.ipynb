```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
dataset=pd.read_csv('Algerian_forest_fires_dataset_UPDATE.csv' ,header=1)
```

```python
dataset.head()
```

```python
dataset.info()
```

```python
dataset[dataset.isnull().any(axis=1)]
```

```python
dataset.loc[:122,"Region"]=0
dataset.loc[122:,"Region"]=1
df=dataset
```

```python
df.info()
```

```python
df[['Region']]=df[['Region']].astype(int)
```

```python
df
```

```python
df.isnull().sum()
```

```python
## Removing the null values
df=df.dropna().reset_index(drop=True)
```

```python
df
```

```python
df.isnull().sum()
```

```python
df.iloc[[122]]
```

```python
##remove the 122nd row
df=df.drop(122).reset_index(drop=True)
```

```python
df.iloc[[122]]
```

```python
df.columns
```

```python
## fix spaces in columns names
df.columns=df.columns.str.strip()
df.columns
```

```python
df.info()
```

#### Changing the required columns as integer data type

```python
df[['day','month','year','Temperature','RH','Ws']]=df[['day','month','year','Temperature','RH','Ws']].astype(int)
```

```python
df.info()
```

#### Changing the other columns to float data datatype


```python
objects=[features for features in df.columns if df[features].dtypes=='O']
```

```python
objects
```

```python
for i in objects:
    if i!='Classes':
        df[i]=df[i].astype(float)
```

```python
df.info()
```

```python
df.describe()
```

```python
## Let save the cleaned dataset
df.to_csv('Algerian_forest_fires_cleaned_dataset_Vishal.csv',index=False)
```

```python
df['Classes']=np.where(df['Classes'].str.contains('not fire'),'not fire','fire')
```

```python
## Monthly Fire Analysis
dftemp1=df.loc[df['Region']==1]
plt.subplots(figsize=(13,6))

sns.set_style('whitegrid')

sns.countplot(x='month',hue='Classes',data=dftemp1)

plt.ylabel('Number of Fires')
plt.xlabel('Months')
plt.title("Fire Analysis of Sidi- Bel Regions",weight='bold')
```

```python
## Monthly Fire Analysis
dftemp2=df.loc[df['Region']==0]

plt.subplots(figsize=(13,6))


sns.countplot(x='month',hue='Classes',data=dftemp2,palette={'not fire': 'green','fire': 'red'})

plt.ylabel('Number of Fires')
plt.xlabel('Months')
plt.title("Fire Analysis of  Brjaia- Regions",weight='bold')
```
